## Prepare the data for the model

```python
teams = training_df['home_team'].unique()
teams = pd.DataFrame(teams, columns=['team'])
teams['i'] = teams.index

training_df = pd.merge(training_df, teams, left_on='home_team', right_on='team', how='left')
training_df = training_df.rename(columns={'i': 'i_home'}).drop('team', 1)
training_df = pd.merge(training_df, teams, left_on='away_team', right_on='team', how='left')
training_df = training_df.rename(columns={'i': 'i_away'}).drop('team', 1)

observed_home_goals = training_df['home_score'].values
observed_away_goals = training_df['away_score'].values

home_team = training_df['i_home'].values
away_team = training_df['i_away'].values

num_teams = len(training_df['i_home'].drop_duplicates())
num_games = len(home_team)
```

```python
g = training_df.groupby('i_away')
att_starting_points = np.log(g.away_score.mean())
g = training_df.groupby('i_home')
def_starting_points = -np.log(g.away_score.mean())
```

## Build the model

```python
with pm.Model() as model:
    # global model parameters
    home = pm.Normal('home', 0, 0.0001)
    tau_att = pm.Gamma('tau_att', .1, .1)
    tau_def = pm.Gamma('tau_def', .1, .1)
    intercept = pm.Normal('intercept', 0, 0.0001)
    
    # team specific parameters
    atts_star = pm.Normal('atts_star', mu=0, tau=tau_att, shape=num_teams)
    defs_star = pm.Normal('defs_star', mu=0, tau=tau_def, shape=num_teams)
    
    #
    atts = pm.Deterministic('atts', atts_star - tt.mean(atts_star))
    defs = pm.Deterministic('defs', defs_star - tt.mean(defs_star))
    
    home_theta = tt.exp(intercept + home + atts[home_team] + defs[away_team])
    away_theta = tt.exp(intercept + atts[away_team] + defs[home_team])
    
    # likelihood of observed data
    home_goals = pm.Poisson('home_goals', mu=home_theta, observed=observed_home_goals)
    away_goals = pm.Poisson('away_goals', mu=away_theta, observed=observed_away_goals)
```

## Fit the model

```python
with model:
    
    start = pm.find_MAP()
    step = pm.NUTS(state=start)
    trace = pm.sample(2000, step, start=start, njobs=1)
    
pm.traceplot(trace)
```

# Utility Functions

```python
def home_goals(intercept, home_advantage, attack_score, defence_score):
    try:
        mu =  np.exp(intercept * home_advantage * attack_score * defence_score)
    except TypeError as e:
        print e
        print intercept, home_advantage, attack_score, defence_score
        raise
    #print mu
    if mu <= 0:
        #print mu
        mu = 0.0000001
    return poisson(mu).rvs(1)

def away_goals(intercept, attack_score, defence_score):
    try:
        lambda_param = np.exp(intercept * attack_score * defence_score)
    except TypeError as e:
        print e
        print intercept, attack_score, defence_score
        raise
    if lambda_param <= 0:
        lambda_param = 0.0000001
    return poisson(lambda_param).rvs(1)
```

```python
def result_after_x_iters(intercept_val, home_val, atts_val_home, defs_val_home, atts_val_away, defs_val_away, x):
    result_list = []
    for iteration in xrange(x):
        
        home_score = home_goals(intercept_val, home_val, atts_val_home, defs_val_away)
        away_score = away_goals(intercept_val, atts_val_away, defs_val_home)
        
        #print "Home:{home} v {away}: Away".format(home=home_score, away=away_score)
    if home_score > away_score:
        result_list.append('H')
    if home_score == away_score:
        result_list.append('D')
    if home_score < away_score:
        result_list.append('L')
    return mode(result_list)[0][0]
```

```python
def result_of_fixture(home_team, away_team, atts_df, defs_df, intercept_df, home_df):
    assert len(atts_df) == len(defs_df) == len(home_df) == len(intercept_df)    
    result_list = []
    burn = 5
    atts_df = atts_df.iloc[burn::]
    defs_df = defs_df.iloc[burn::]
    intercept_df = intercept_df.iloc[burn::]
    home_df = home_df.iloc[burn::]
    for iteration_num in xrange(50):            
        intercept_val = intercept_df[0].iloc[iteration_num]
        home_val = home_df[0].iloc[iteration_num]
        atts_val_home = atts_df[home_team].iloc[iteration_num]
        defs_val_home = defs_df[home_team].iloc[iteration_num]
        atts_val_away = atts_df[away_team].iloc[iteration_num]
        defs_val_away = defs_df[away_team].iloc[iteration_num]
        
        expected_results = result_after_x_iters(intercept_val=intercept_val,
                                                home_val=home_val, 
                                             atts_val_home=atts_val_home,
                                             atts_val_away=atts_val_away,
                                             defs_val_home=defs_val_home,
                                             defs_val_away=defs_val_away,
                                             x=10)
        result_list.append(expected_results)
    return  result_list
```

```python
def plot_confusion_matrix(cm, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(prediction_df.actual_result.unique()))
    plt.xticks(tick_marks, prediction_df.actual_result.unique(), rotation=45)
    plt.yticks(tick_marks, prediction_df.actual_result.unique())
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```
